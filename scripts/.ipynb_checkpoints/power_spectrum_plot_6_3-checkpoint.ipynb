```python
def average_power_spectrum(era_data,forecast_data,speedy_data=None):
    time_step = 1/4.0#1/14540
    length_of_data = len(era_data)
    power_spectra_era = np.zeros((length_of_data))
    power_spectra_forecast = np.zeros((length_of_data))
    power_spectra_speedy = np.zeros((length_of_data))

    hamming = np.hanning(length_of_data)
#     for i in range(np.shape(era_data)[1]):
#         for j in range(np.shape(era_data)[2]):

    fft_era = np.fft.fft(era_data[0:length_of_data]*hamming)
    fft_forecast = np.fft.fft(forecast_data[0:length_of_data]*hamming)

    power_spectra_era[:] = np.abs(fft_era)**2.0
    power_spectra_forecast[:]= np.abs(fft_forecast)**2.0

    if speedy_data is not None:
        fft_speedy = np.fft.fft(speedy_data[0:length_of_data]*hamming)
        power_spectra_speedy[:] = np.abs(fft_speedy)**2.0

    print(np.shape(power_spectra_speedy),'power_spectra_speedy shape')
    freq = np.fft.fftfreq(len(speedy_data[0:length_of_data]),time_step)

    idx = np.argsort(freq)
    idx = idx[int(len(idx)/2)::]

   #freq = 4*(freq/14540)

    averaged_power_spectrum_era = np.average(power_spectra_era)
    averaged_power_spectrum_hybrid = np.average(power_spectra_forecast)
    averaged_power_spectrum_speedy = np.average(power_spectra_speedy)

    return averaged_power_spectrum_era, averaged_power_spectrum_hybrid, averaged_power_spectrum_speedy, freq, idx

```

```python
def power_spectra_mitchell(data):
    time_step = 1 /4
    power_spectra_era = np.zeros((np.shape(data)))

    hamming = np.hamming(np.shape(data)[0])

    fft_data = np.fft.fft(data*hamming)
    power_spectra_data = np.abs(fft_data)**2.0

    freq = np.fft.fftfreq(len(data),time_step)

    idx = np.argsort(freq)
    idx = idx[int(len(idx)/2)::]

    return power_spectra_data, freq, idx
```

```python
time_series_data = hybrid_1_9_1_9_at_sahara

density, f, idx = power_spectra_mitchell(time_series_data)

ax1.semilogy(f[:-1*np.shape(idx)[0]], density[:-1*np.shape(idx)[0]])
```
